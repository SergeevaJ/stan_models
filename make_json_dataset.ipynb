```python
#add necessary columns addl, ss, rate, ii -zeros
data["addl"]=np.zeros(data.shape[0], dtype=int)
data["ss"]=np.zeros(data.shape[0],dtype=int)
data["rate"]=np.zeros(data.shape[0],dtype=int)
data["ii"]=np.zeros(data.shape[0], dtype=int)
```

```python
# change cmt for adm events
data["CMT"]=2
data.loc[data["EVID"]==1, "CMT"]=np.zeros(9)+1
```

```python
data.rename(columns={'TIME':'time', 'AMT': 'amt', 'EVID':'evid', 'CMT':'cmt', 'DV': 'cObs' }, inplace=True)
```

```python
#first dict of nt, nObs, nSubjects, iObs, start, end, then our df
df=data[['cmt', 'evid','addl', 'ss','amt', 'time', 'rate', 'ii', 'cObs']]
```

```python
#change cObs to have only observation data
```

```python
## add to json
d={"nt" : nt, "nObs":int(nObs), "nSubjects": int(nSubjects), "iObs":iObs.tolist(), "start":start.tolist(), "end":end.tolist()}
```

```python
for col in ['cmt', 'evid','addl', 'ss','amt', 'time', 'rate', 'ii']:
    d[col]=data[col]
```

```python
ind=iObs-1
d['cObs']=data.loc[ind,'cObs']
```

```python
for val in d.values():
  print(type(val))
```

```python
st.write_stan_json(path='datapkcmt2.json', data=d)
```
